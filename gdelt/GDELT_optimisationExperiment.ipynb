```python
import pandas as pd
import gdelt
```

Using GDELTPyR library from https://github.com/linwoodc3/gdeltPyR/tree/proxy_fix

```python
# Version 2 queries
gd2 = gdelt.gdelt(version=2)
```

```python
%%time
resultsMentions = gd2.Search(['2020 12 14'],
                             table='events',
                             coverage=True,
                             output='json')
```

```python
# df.memory_usage(deep=True)/(1024*1024)
```

```python
df = pd.read_json(resultsMentions)
df_IN = df.loc[df['ActionGeo_CountryCode'] == 'IN']
dedup_IN = df_IN.drop_duplicates(subset=['SOURCEURL'])
```

```python
dedup_IN
```

```python
%%time
# Download GDELT Data
resultsEventsTrans = gd2.Search(['2020 12 14'],
                                table='events',
                                translation=True,
                                coverage=True,
                                output='json')
# Read JSON
transdf = pd.read_json(resultsEventsTrans)
# 
transdf_IN = transdf.loc[transdf['ActionGeo_CountryCode'] == 'IN']

transdf_IN_dedup = transdf_IN.drop_duplicates(subset=['SOURCEURL'])
```

```python
transdf_IN_dedup
```

```python
data = dedup_IN.append(transdf_IN_dedup)
```

```python
data.columns
```

```python
data.get(['GLOBALEVENTID',
          'DATEADDED',
          'SOURCEURL',
          'EventCode',
          'CAMEOCodeDescription',
          'QuadClass',
          'GoldsteinScale',
          'AvgTone',
          'ActionGeo_CountryCode',
          'ActionGeo_FullName', 
          'ActionGeo_Lat', 
          'ActionGeo_Long'])
```

```python
extractedEventsData = data.get(['GLOBALEVENTID',
          'DATEADDED',
          'SOURCEURL',
          'EventCode',
          'CAMEOCodeDescription',
          'QuadClass',
          'GoldsteinScale',
          'AvgTone',
          'ActionGeo_CountryCode',
          'ActionGeo_FullName', 
          'ActionGeo_Lat', 
          'ActionGeo_Long'])
```

```python
extractedEventsData.reset_index(drop=True, inplace=True)
```
