```python
def extract(url):
    article = newspaper.Article(url)
    article.download()
    try:
        article.parse()
        article.nlp()
        
        title = article.title
        author = article.authors
        
        keywords = article.keywords
        summary = article.summary
        
        text = article.text
        topImage = article.top_image
        images = article.images
        videos = article.movies
        
        html = ""
        html = article.html
        
        if html == "":
            htmldata = getdata(url) 
            soup = BeautifulSoup(htmldata, 'html5lib') 
            return title, author, keywords, summary, text, topImage, images, videos, soup
        else:
            return title, author, keywords, summary, text, topImage, images, videos, html
    
    except:
        
        return "", "", "", "", "", "", "", "", ""
```

```python
import requests 
import pandas as pd
import html5lib
from bs4 import BeautifulSoup 
  
# link for extract html data 
def getdata(url): 
    r = requests.get(url) 
    return r.text 
```

```python
htmldata = getdata("https://www.finanzen.ch/nachrichten/aktien/dachser-air-&-sea-logistics-certified-for-pharmaceutical-shipments-on-three-continents-1028650203") 
soup = BeautifulSoup(htmldata, 'html5lib') 
data = '' 
for data in soup.find_all("p"): 
    print(data.get_text())
```

```python
count = 0
for url in df['SOURCEURL']:
    req = requests.head(url)
    if req.status_code == requests.codes.ok:
        count += 1
print("Websites returning 200 OK ", count)
print("NOT 200 OK ", len(df['SOURCEURL']) - count)
```

```python
def exists(url):
    try:
        req = requests.head(url, timeout=5)
        if req.status_code == 404:
            return 0, req.status_code
        elif req.status_code == 200:
            return 1, req.status_code
        else:
            return 2, req.status_code
    except:
        return 3, req.status_code
```

```python
%%time
pool = ThreadPool(1000)
existsArray = pool.map(exists, df['SOURCEURL'])
pool.close()
pool.join()
```
